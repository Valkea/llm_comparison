```python
import json
import pandas as pd
```

```python
with open('data/test_freq.json', 'r') as f:
    data = json.load(f)
```

```python
intros = []
for k in data.keys():
    v = data[k]['content'][0]['message']
    print(v)
    intros.append(v)
    if len(intros) >= 100:
        break    
```

```python
len(intros)
```

```python
intro_sequences = []
for k in data.keys():
    
    dialogs = []
    for dialog in data[k]['content']:
        a = dialog['agent']
        v = dialog['message']
        print("######", a, v)
        dialogs.append(f"{a}: {v}")

        # Break on second question of the "user"
        if len(dialogs) > 1:
            if v.find('?') > 0 and a == 'agent_1':
                break
                
    if len(dialogs) <= 3:
        dialogs.append("\n")
        intro_sequences.append("\n".join(dialogs))
        # intro_sequences.append(dialogs)
        print("WE TAKE THIS ONE")
    print("****************")
    if len(intro_sequences) >= 50:
        break    
```

```python
len(intro_sequences)
```

```python
datapd = pd.DataFrame(intro_sequences)
```

```python
from nltk.lm import MLE
lm = MLE(2)
```

```python
from nltk.lm.preprocessing import padded_everygram_pipeline
text = "Blop pour voir"
train, vocab = padded_everygram_pipeline(2, text)
lm.fit(train, vocab)
print(lm.vocab.lookup(text[0]))
print(len(lm.vocab))
test = [('Blop'), ('pour')]
lm.perplexity(test)
```
